```python


import pandas as pd
import numpy as np

df = pd.read_csv("1_test_gop_sentiment_binary.csv")
conf_true = []
for _, row in df.iterrows():
    if row['gold_label'] == 0:
        conf_true.append(row['conf0'])
    else:
        conf_true.append(row['conf1'])
df2 = pd.DataFrame(conf_true)
df2.hist()
print()
```
